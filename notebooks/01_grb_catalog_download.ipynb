```python
from pathlib import Path
import pandas as pd

BASE = Path.home() / "thesis"
TEX = BASE / "outputs" / "tex"
TAB = BASE / "outputs" / "tables"
TEX.mkdir(parents=True, exist_ok=True)

# ---------- GRB ----------
g = pd.read_csv(TAB/"grb_summary_stats.csv").iloc[0]
grb_tex = (
    f"\\newcommand{{\\GRBTotal}}{{{int(g['N_total'])}}}\n"
    f"\\newcommand{{\\GRBShort}}{{{int(g['N_short_T90_lt_2s'])}}}\n"
    f"\\newcommand{{\\GRBLong}}{{{int(g['N_long_T90_ge_2s'])}}}\n"
    f"\\newcommand{{\\GRBShortFrac}}{{{float(g['short_fraction']):.3f}}}\n"
    f"\\newcommand{{\\GRBTNinetyMed}}{{{float(g['T90_median_s']):.3f}}}\n"
    f"\\newcommand{{\\GRBTNinetyP16}}{{{float(g['T90_16pct_s']):.3f}}}\n"
    f"\\newcommand{{\\GRBTNinetyP84}}{{{float(g['T90_84pct_s']):.3f}}}\n"
)
(TEX/"grb_numbers.tex").write_text(grb_tex)

# ---------- AGN ----------
a = pd.read_csv(TAB/"agn_summary_stats.csv").iloc[0]
agn_tex = (
    f"\\newcommand{{\\AGNTot}}{{{int(a['N_total_sources'])}}}\n"
    f"\\newcommand{{\\AGNBLL}}{{{int(a['N_BLL'])}}}\n"
    f"\\newcommand{{\\AGNFSRQ}}{{{int(a['N_FSRQ'])}}}\n"
    f"\\newcommand{{\\AGNGammaMedBLL}}{{{float(a['Gamma_median_BLL']):.3f}}}\n"
    f"\\newcommand{{\\AGNGammaMedFSRQ}}{{{float(a['Gamma_median_FSRQ']):.3f}}}\n"
)
# only if present in your CSV
if "Variability_median_BLL" in a.index:
    agn_tex += f"\\newcommand{{\\AGNVarMedBLL}}{{{float(a['Variability_median_BLL']):.3f}}}\n"
if "Variability_median_FSRQ" in a.index:
    agn_tex += f"\\newcommand{{\\AGNVarMedFSRQ}}{{{float(a['Variability_median_FSRQ']):.3f}}}\n"
(TEX/"agn_numbers.tex").write_text(agn_tex)

# ---------- CR ----------
c = pd.read_csv(TAB/"cr_proton_modulation_summary.csv").iloc[0]
cr_tex = (
    f"\\newcommand{{\\CRRigMin}}{{{float(c['rigidity_min_GV']):.2f}}}\n"
    f"\\newcommand{{\\CRRigMax}}{{{float(c['rigidity_max_GV']):.2f}}}\n"
    f"\\newcommand{{\\CRNPoints}}{{{int(c['N_points'])}}}\n"
    f"\\newcommand{{\\CRFluxMean}}{{{float(c['mean_flux']):.2f}}}\n"
    f"\\newcommand{{\\CRFluxMin}}{{{float(c['min_flux']):.1f}}}\n"
    f"\\newcommand{{\\CRFluxMax}}{{{float(c['max_flux']):.1f}}}\n"
    f"\\newcommand{{\\CRAmpFrac}}{{{float(c['fractional_amplitude_(max-min)/mean']):.3f}}}\n"
)
(TEX/"cr_numbers.tex").write_text(cr_tex)

print("Wrote:", TEX/"grb_numbers.tex")
print("Wrote:", TEX/"agn_numbers.tex")
print("Wrote:", TEX/"cr_numbers.tex")
```

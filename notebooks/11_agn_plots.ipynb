```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

BASE = Path.home()/ "thesis"
agn = pd.read_csv(BASE/"data/processed/agn/agn_4lac_clean.csv", low_memory=False)

cls = agn["CLASS"].astype(str).str.lower()
gamma = pd.to_numeric(agn["PL_Index"], errors="coerce")

bll = gamma[cls.str.contains("bll")].dropna()
fsrq = gamma[cls.str.contains("fsrq")].dropna()

plt.figure(figsize=(7,4))
bins = np.linspace(1.2, 3.5, 45)
plt.hist(bll,  bins=bins, histtype="step", linewidth=2, label="BL Lac")
plt.hist(fsrq, bins=bins, histtype="step", linewidth=2, label="FSRQ")

plt.xlabel(r"Photon index $\Gamma$ (PL\_Index)")
plt.ylabel("Number of sources")
plt.title("4LAC: Photon Index Distributions")
plt.legend()
plt.tight_layout()

out = BASE/"figures/Fig03_AGN_photon_index_BLL_FSRQ.png"
plt.savefig(out, dpi=300)
print("Saved:", out)
plt.show()
```

```python
var = pd.to_numeric(agn["Frac_Variability"], errors="coerce")
m = np.isfinite(gamma) & np.isfinite(var)

plt.figure(figsize=(6,5))
plt.scatter(gamma[m & cls.str.contains("bll")],  var[m & cls.str.contains("bll")],  s=6, label="BL Lac")
plt.scatter(gamma[m & cls.str.contains("fsrq")], var[m & cls.str.contains("fsrq")], s=6, label="FSRQ")

plt.xlabel(r"Photon index $\Gamma$")
plt.ylabel("Frac_Variability")
plt.title("4LAC: Variability vs Spectral Slope")
plt.legend()
plt.tight_layout()

out = BASE/"figures/Fig04_AGN_var_vs_index.png"
plt.savefig(out, dpi=300)
print("Saved:", out)
plt.show()
```

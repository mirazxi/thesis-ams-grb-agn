```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

BASE = Path.home()/ "thesis"
cr = pd.read_csv(BASE/"data/processed/cr/ams02_proton_daily.csv")
cr["date"] = pd.to_datetime(cr["date"], errors="coerce")
cr = cr.dropna(subset=["date","flux","rigidity_min_GV","rigidity_max_GV"])

rmin, rmax = 3.29, 3.64
sel = cr[np.isclose(cr["rigidity_min_GV"], rmin) & np.isclose(cr["rigidity_max_GV"], rmax)].copy()
sel = sel.sort_values("date")

# total error if present
yerr = None
if {"err_stat","err_time","err_sys"}.issubset(sel.columns):
    yerr = np.sqrt(sel["err_stat"]**2 + sel["err_time"]**2 + sel["err_sys"]**2)

plt.figure(figsize=(10,4))
if yerr is not None:
    plt.errorbar(sel["date"], sel["flux"], yerr=yerr, fmt=".", capsize=2)
else:
    plt.plot(sel["date"], sel["flux"], ".")

plt.xlabel("Date")
plt.ylabel(r"Proton flux [$\mathrm{m^{-2}\,sr^{-1}\,s^{-1}\,GV^{-1}}$]")
plt.title(f"AMS-02 Proton Flux vs Time ({rmin}-{rmax} GV)")
plt.tight_layout()

out = BASE/"figures/Fig05_CR_proton_flux_time_3p29_3p64GV.png"
plt.savefig(out, dpi=300)
print("Saved:", out)
plt.show()
```

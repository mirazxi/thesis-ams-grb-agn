```python
from gdt.missions.fermi.gbm.catalogs import BurstCatalog
import pandas as pd
from astropy.table import Table
from pathlib import Path

cat = BurstCatalog()

# Find an astropy Table inside the object (version-proof)
tbl = None
for name in dir(cat):
    if name.startswith("_"):
        continue
    try:
        val = getattr(cat, name)
    except Exception:
        continue
    if isinstance(val, Table):
        tbl = val
        print("Found astropy Table at attribute:", name)
        break

# If it's stored as a private attribute, look there too
if tbl is None:
    for name in ["_table", "_catalog", "catalog", "data", "_data"]:
        if hasattr(cat, name):
            val = getattr(cat, name)
            if isinstance(val, Table):
                tbl = val
                print("Found astropy Table at attribute:", name)
                break

if tbl is None:
    raise RuntimeError("Couldn't find the catalog table inside BurstCatalog. We'll inspect cat next.")

df = tbl.to_pandas()

out = Path.home() / "thesis" / "data" / "grb" / "fermigbrst.csv"
out.parent.mkdir(parents=True, exist_ok=True)
df.to_csv(out, index=False)

print("Saved:", out)
print("Rows:", len(df), "Cols:", len(df.columns))
```

```python
from pathlib import Path
import pandas as pd

root = Path.home() / "thesis" / "data" / "cr" / "AMS_timedependent_datasets"
outdir = Path.home() / "thesis" / "data" / "cr" / "processed"
outdir.mkdir(parents=True, exist_ok=True)

# candidate table files
cand = [p for p in root.rglob("*") if p.is_file() and p.suffix.lower() in [".csv", ".txt", ".dat", ".tsv"]]

loaded = []
failed = []

def load_table(path: Path) -> pd.DataFrame:
    suf = path.suffix.lower()
    if suf == ".csv":
        return pd.read_csv(path)
    if suf == ".tsv":
        return pd.read_csv(path, sep="\t")
    # txt/dat: usually whitespace-delimited with comments
    return pd.read_csv(path, delim_whitespace=True, comment="#", engine="python")

for p in cand:
    try:
        df = load_table(p)
        # drop completely empty columns
        df = df.dropna(axis=1, how="all")
        # save with safe filename
        safe = p.relative_to(root).as_posix().replace("/", "__")
        out = outdir / f"AMS02__{safe}.csv"
        df.to_csv(out, index=False)
        loaded.append((p, out, df.shape))
    except Exception as e:
        failed.append((p, str(e)))

print("Total candidate files:", len(cand))
print("Loaded:", len(loaded))
print("Failed:", len(failed))

# show a few loaded examples
for p, out, shp in loaded[:10]:
    print("OK:", p.name, "->", out.name, "shape:", shp)

# if something failed, show first 5 errors
for p, err in failed[:5]:
    print("FAIL:", p.name, "error:", err[:120])
```

```python
from pathlib import Path
import pandas as pd

root = Path.home() / "thesis" / "data" / "cr" / "AMS_timedependent_datasets"
outdir = Path.home() / "thesis" / "data" / "cr" / "processed"
outdir.mkdir(parents=True, exist_ok=True)

cand = [p for p in root.rglob("*") if p.is_file() and p.suffix.lower() in [".csv", ".txt", ".dat", ".tsv"]]

def is_readme(p: Path) -> bool:
    name = p.name.lower()
    return ("readme" in name) or ("license" in name) or ("citation" in name)

cand = [p for p in cand if not is_readme(p)]  # <-- skip README-like files

loaded, failed = [], []

def load_table(path: Path) -> pd.DataFrame:
    suf = path.suffix.lower()
    if suf == ".csv":
        return pd.read_csv(path)
    if suf == ".tsv":
        return pd.read_csv(path, sep="\t")
    # txt/dat: whitespace tables
    return pd.read_csv(path, sep=r"\s+", comment="#", engine="python")

for p in cand:
    try:
        df = load_table(p).dropna(axis=1, how="all")
        safe = p.relative_to(root).as_posix().replace("/", "__")
        out = outdir / f"AMS02__{safe}.csv"
        df.to_csv(out, index=False)
        loaded.append((p.name, out.name, df.shape))
    except Exception as e:
        failed.append((p.name, str(e)))

print("Candidate files:", len(cand))
print("Loaded:", len(loaded))
print("Failed:", len(failed))
if failed:
    print("First failure:", failed[0])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

p = pd.read_csv(Path.home()/"thesis/data/cr/core/ams02_proton_daily.csv")
print("Columns:", list(p.columns))

# try to pick sensible columns automatically
def pick(df, keys):
    for k in keys:
        for c in df.columns:
            if k in c.lower():
                return c
    return None

time_col = pick(p, ["bartels", "rotation", "time", "date", "mjd"])
flux_col = pick(p, ["flux", "intensity"])
err_col  = pick(p, ["err", "error", "sigma", "unc"])

print("Using:", time_col, flux_col, err_col)

plt.figure()
if err_col:
    plt.errorbar(p[time_col], p[flux_col], yerr=p[err_col], fmt=".", capsize=2)
else:
    plt.plot(p[time_col], p[flux_col], ".")
plt.xlabel(time_col)
plt.ylabel(flux_col)
plt.title("AMS-02 Proton Flux (time-dependent)")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import pandas as pd
from pathlib import Path
df = pd.read_csv(Path.home()/"thesis/data/cr/core/ams02_proton_daily.csv")

# Rename columns (easier to use)
date_col = "date YYYY-MM-DD"
rmin_col = "rigidity_min GV"
rmax_col = "rigidity_max GV"
flux_col = "proton_flux m^-2sr^-1s^-1GV^-1"
err_stat_col = "proton_flux_error_statistical m^-2sr^-1s^-1GV^-1"
err_time_col = "proton_flux_error_timedependent m^-2sr^-1s^-1GV^-1"
err_sys_col  = "proton_flux_error_systematic_total m^-2sr^-1s^-1GV^-1"

# Parse date
df[date_col] = pd.to_datetime(df[date_col], errors="coerce")
df = df.dropna(subset=[date_col])

# Choose ONE rigidity bin (edit these numbers anytime)
target_rmin, target_rmax = 2.0, 3.0
df = pd.read_csv(Path.home()/"thesis/data/cr/core/ams02_proton_daily.csv")

bins = (df[["rigidity_min GV","rigidity_max GV"]]
        .drop_duplicates()
        .sort_values(["rigidity_min GV","rigidity_max GV"])
        .reset_index(drop=True))

print("Number of bins:", len(bins))
display(bins.head(20))          # first 20 bins
display(bins.tail(20))          # last 20 bins


sel = df[(df[rmin_col] == target_rmin) & (df[rmax_col] == target_rmax)].copy()
print("Selected rows:", len(sel), "for rigidity bin:", target_rmin, "-", target_rmax, "GV")

# Total uncertainty (stat + time-dependent + systematic) in quadrature (optional)
has_all_err = all(c in sel.columns for c in [err_stat_col, err_time_col, err_sys_col])
if has_all_err:
    sel["err_total"] = np.sqrt(sel[err_stat_col]**2 + sel[err_time_col]**2 + sel[err_sys_col]**2)
    yerr = sel["err_total"]
    yerr_label = "total error (stattimesys)"
else:
    yerr = sel[err_stat_col] if err_stat_col in sel.columns else None
    yerr_label = "statistical error" if yerr is not None else None

# Sort by date
sel = sel.sort_values(date_col)

# Plot
plt.figure(figsize=(10,4))
if yerr is not None:
    plt.errorbar(sel[date_col], sel[flux_col], yerr=yerr, fmt=".", capsize=2)
else:
    plt.plot(sel[date_col], sel[flux_col], ".")

plt.xlabel("Date")
plt.ylabel(r"Proton flux [$\mathrm{m^{-2}\,sr^{-1}\,s^{-1}\,GV^{-1}}$]")
plt.title(f"AMS-02 Proton Flux vs Time ({target_rmin}-{target_rmax} GV), with {yerr_label or 'no error bars'}")
plt.tight_layout()
plt.show()
```
